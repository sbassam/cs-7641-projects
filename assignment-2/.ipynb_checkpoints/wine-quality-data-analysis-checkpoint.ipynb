```python
df1 = pd.read_csv('data/winequality-red.csv', sep = ';')
```

```python
df1 ['red'] = 1
df1 ['white'] = 0
```

```python
df2 = pd.read_csv('data/winequality-white.csv', sep = ';')
```

```python
df2 ['red'] = 0
df2 ['white'] = 1
```

```python
wine_quality = pd.concat([df1, df2], axis = 0)
```

```python
wine_quality[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']].values
```

```python
wine_quality.head()
```

```python
distance_metric_arr = np.array(['euclidean', 'manhattan', 'chebyshev', 'minkowski', 
                                  'wminkowski', 'seuclidean', 'mahalanobis'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
features = wine_quality[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'red', 'white']].values
target = wine_quality[['quality']]
features_train, features_test, labels_train, labels_test = train_test_split(features, target, test_size=0.3)
```

```python
clf = DecisionTreeClassifier()
clf.fit(features_train, labels_train)
pred = clf.predict(features_test)
```

```python
compare = pd.DataFrame(labels_test)
compare['predicted_quality'] = pred
error = compare[compare['quality'] !=compare['predicted_quality']]
compare.shape
```

```python
ax = sns.scatterplot(x="quality", y="predicted_quality", data=compare)
```

```python
y = labels_test
fig, ax = plt.subplots()
ax.scatter(y, pred)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

```python
dt_score = clf.score(features_test, labels_test)
dt_precision = precision_score(labels_test, pred, average='weighted')
dt_recall = recall_score(labels_test, pred, average='weighted')
dt_f1 = f1_score(labels_test, pred, average='weighted')
print (dt_score, dt_precision, dt_recall, dt_f1)
```

```python
scaler = StandardScaler()
scaler.fit(features_train)
StandardScaler(copy=True, with_mean=True, with_std=True)
features_train = scaler.transform(features_train)
features_test = scaler.transform(features_test)
```

```python
clf = DecisionTreeClassifier()
clf.fit(features_train, labels_train)
pred = clf.predict(features_test)
```

```python
dt_score = accuracy_score(labels_test, pred)
dt_precision = precision_score(labels_test, pred, average='weighted')
dt_recall = recall_score(labels_test, pred, average='weighted')
dt_f1 = f1_score(labels_test, pred, average='weighted')
print (dt_score, dt_precision, dt_recall, dt_f1)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'max_depth':np.arange(1, 50, 5), 'max_leaf_nodes': np.arange(2, 50, 5)}
gc_clf = GridSearchCV(clf, parameters, cv=5)
gc_clf.fit(features_train, labels_train)
```

```python
gc_clf.best_params_
```

```python
means = gc_clf.cv_results_['mean_test_score']
stds = gc_clf.cv_results_['std_test_score']
```
