```python
import mlrose
import numpy as np
import matplotlib.pyplot as plt
import time
import pandas as pd
import seaborn as sns
```

```python
# Four Peaks - RHC
from mlrose import FourPeaks, RHCRunner, SARunner, GARunner, MIMICRunner, DiscreteOpt
fitness = FourPeaks()
problem = DiscreteOpt(length=100, fitness_fn=fitness, maximize=True, max_val=2)
rhc_tuning = pd.DataFrame()

## run 5 times
for i in range(5):
    rhc = RHCRunner(problem,
                    experiment_name='rhc_len_10_run_'+str(i),
                    output_directory='out/',
                    seed = i,
                    iteration_list=2 ** np.arange(16),
                    restart_list=[1, 5, 10, 20, 50, 100],
                    max_attempts=5000,
                    generate_curves=True)
    
    df_str, rhc_run_curves = rhc.run()
    df_str['run_number'] = i
    rhc_tuning = rhc_tuning.append(df_str)
```

```python
# group by restart and Iteration
rhc_grouped = rhc_tuning[['Fitness', 'Restarts', 'Iteration', 'run_number']]
rhc_grouped = rhc_grouped.groupby(['Restarts', 'Iteration'])
```

```python
# Take the mean of runs
rhc_grouped = rhc_grouped.agg({'Fitness': np.mean})
```

```python
# https://scentellegher.github.io/programming/2017/07/15/pandas-groupby-multiple-columns-plot.html
fig, ax = plt.subplots(figsize=(15,7))
plt.xscale('log', basex=2)
plt.ylabel('Mean Fitness over 5 runs', fontsize=14)
plt.title('RHC Tuning - Mean Fitness vs Iteration for various restarts', fontsize=14)
# use unstack()
rhc_grouped.groupby(['Iteration','Restarts']).mean()['Fitness'].unstack().plot(ax=ax)
```

```python
# Four Peaks - SA
from mlrose import FourPeaks, RHCRunner, SARunner, GARunner, MIMICRunner, DiscreteOpt
fitness = FourPeaks()
problem = DiscreteOpt(length=100, fitness_fn=fitness, maximize=True, max_val=2)
sa_tuning = pd.DataFrame()
## run 5 times
for i in range(5):
    sa = SARunner(problem,
                experiment_name='sa_n',
                seed=i,
                iteration_list=2 ** np.arange(16),
                temperature_list=[1, 10, 50, 100, 150, 250, 300],
                max_attempts=5000,
                generate_curves=True)
    sa_str, sa_run_curves = sa.run()
    sa_str['run_number'] = i
    sa_tuning = sa_tuning.append(sa_str)

```

```python
# Temperature is actually a GeomDecay object. get the actual temperature:
sa_tuning['Temperature'] = sa_tuning['Temperature'].apply(lambda x: x.init_temp)
# group by Temperature and Iteration
sa_grouped = sa_tuning[['Fitness', 'Temperature', 'Iteration', 'run_number']]
sa_grouped = sa_grouped.groupby(['Temperature', 'Iteration'])
```

```python
# Take the mean of runs
sa_grouped = sa_grouped.agg({'Fitness': np.mean})
```

```python
# https://scentellegher.github.io/programming/2017/07/15/pandas-groupby-multiple-columns-plot.html
fig, ax = plt.subplots(figsize=(15,7))
plt.xscale('log', basex=2)
plt.ylabel('Mean Fitness over 5 runs', fontsize=14)
plt.title('SA Tuning - Mean Fitness vs Iteration for various temperatures', fontsize=14)
# use unstack()
sa_grouped.groupby(['Iteration','Temperature']).mean()['Fitness'].unstack().plot(ax=ax)
```
